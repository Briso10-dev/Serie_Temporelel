<h1 style='color:rgb(0, 140, 255)'> Importation des modules </h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h2 style='color:rgb(0, 140, 255); text-align:center'>Calcule de la moyenne,la variance et l'écart-type<h2>

```python
def calculate_statistics(data):
    mean = np.mean(data)
    variance = np.var(data)
    std_dev = np.std(data)
    return mean, variance, std_dev
```

<h2 style='color:rgb(0, 140, 255);text-align:center'> Representation de la série temporelle</h2

```python
def plot_time_series(data):
    x = np.arange(len(data))
    y = data

    plt.plot(x, y, '-o')
    plt.xlabel('Temps')
    plt.ylabel('Valeur(X)')
    plt.title('Série temporelle')
    plt.show()
```

<h2 style='color:rgb(0, 140, 255);text-align:center'> Representation des nuages de points N1,N2,...N8</h2>

```python
def plot_scatter_plots(data):
    num_plots = len(data.columns)
    fig, axs = plt.subplots(num_plots, 1, figsize=(4,20))
    
    for i, col in enumerate(data.columns):
        axs[i].scatter(data.index, data[col])
        axs[i].set_xlabel('Temps')
        axs[i].set_ylabel('Valeur')
        axs[i].set_title(f'Nuage de points {col}')
    
    plt.tight_layout()
    plt.show()
```

<h2 style='color:rgb(0, 140, 255);text-align:center'> Representation de la courbe des autocorrelations<h2>

```python
def plot_autocorrelation(data):
    autocorr = pd.plotting.autocorrelation_plot(data)
    autocorr.set_xlim([1, 50])
    autocorr.set_ylim([-1, 1])
    plt.xlabel('Décalage (k)')
    plt.ylabel('Corrélation(ACF)')
    plt.title('Courbe des autocorrélations')
    plt.show()
```

<h2 style='color:rgb(0, 140, 255);text-align:center'> jeux de données 1 : Serie temporelle lineaire pure <h2>

```python
n = 100  # Nombre d'observations
t = np.arange(1, n+1)
a = 2
b = 3

data_lineaire_trend = a * t + b # Tendance lineaire
noise1 = np.random.normal(0, 1, n) # Bruit aléatoire

#Serie temporelle
data1 = data_lineaire_trend + noise1
```

<h2 style='color:rgb(0, 140, 255);text-align:center'> Affichage des statistiques<h2>

```python
mean, variance, std_dev = calculate_statistics(data1)
print(f"Moyenne: {mean}")
print(f"Variance: {variance}")
print(f"Ecart-type: {std_dev}")
```

<h2 style='color:rgb(0, 140, 255);text-align:center'>Representation de la serie temporelle<h2>

```python
plot_time_series(data1)
```

<h2 style='color:rgb(0, 140, 255);text-align:center'> Representation des nuages de points <h2>

```python
random_data = np.random.rand(len(data1), 8)  
df = pd.DataFrame(random_data, columns=['N1', 'N2', 'N3', 'N4', 'N5', 'N6', 'N7', 'N8'])

plot_scatter_plots(df)
```

<h2 style='color:rgb(0, 140, 255);text-align:center'> Representation de la courbe des autocorrélations <h2>

```python
plot_autocorrelation(data1)
    
print('\n')
```
